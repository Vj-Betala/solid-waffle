```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfMaster = pd.read_csv("archive1/Master.csv")
```

```python
birthCountry = dfMaster["birthCountry"].dropna()
birthCounts = birthCountry.value_counts(normalize=True) * 100
fig, ax = plt.subplots()
ax.pie(birthCounts, labels=birthCounts.keys(),autopct='%1.1f%%')
```
